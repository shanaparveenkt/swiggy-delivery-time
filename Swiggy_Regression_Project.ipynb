```python
#STEP 5 : REMOVE USELESS FEATURES

df=df.drop(columns=['Restaurant_Name'])
df
```

```python
#STEP 6 : DATA PREPROCESSING

#Encode Categorical Variables
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
categorical_cols=['Area','Weather','Traffic']
for col in categorical_cols:
    df[col]=le.fit_transform(df[col])
```

```python
#STEP 7 : FEATURE SELECTION

x=df.drop(columns=['Delivery_Time_min'])
x
```

```python
#Target
y=df['Delivery_Time_min']
y
```

```python
#STEP 8 : TRAIN TEST SPLIT

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
#STEP 9 : FEATURE SCALING

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#STEP 10 : MODEL BUILDING

#Linear Regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#Predictions
y_pred_lr=lr.predict(x_test)
y_pred_lr
```

```python
#STEP 11 : MODEL EVALUATION

#Linear Regression Metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print("Linear Regression")
print("MAE:",mean_absolute_error(y_test,y_pred_lr))
print("RMSE:",np.sqrt(mean_squared_error(y_test,y_pred_lr)))
print("R2 Score:",r2_score(y_test,y_pred_lr))
```

```python
#STEP 12 : ACTUAL VS PREDICTED PLOT

plt.figure()
plt.scatter(y_test,y_pred_lr,alpha=0.7)
plt.plot([y_test.min(),y_test.max()],
         [y_test.min(),y_test.max()])
plt.xlabel("Actual Delivery Time")
plt.ylabel("Predicted Delivery Time")
plt.title("Actual vs Predicted Delivery Time")
plt.show()
```

```python
#STEP 13 : COMPARISON TABLE

comparison_df=pd.DataFrame({"Actual_Delivery_Time": y_test.values,"Predicted_Delivery_Time": y_pred_lr})
comparison_df
```
